```python
import scholarly
import bibtexparser
import requests
```

```python
import pandas as pd
import numpy as np
```

```python
search_query = scholarly.search_pubs_query("ARDS case")
```

```python
results = []
iterator = iter(search_query)

for _ in range(50):
    results.append(next(iterator))
```

```python
print(results[0])
```

```python
def process_entry(entry):
    return {
        "abstract": entry.bib.get("abstract"),
        "author": entry.bib.get("author"),
        "eprint": entry.bib.get("eprint"),
        "title": entry.bib.get("title"),
        "citedby": entry.citedby,
        "url_scholarbib": entry.url_scholarbib,
    }
    
```

```python
process_entry(results[0])
```

```python
df = pd.DataFrame([process_entry(entry) for entry in results])
```

```python
df.dropna()
```

```python
df.to_csv("./ards-case.csv")
```
